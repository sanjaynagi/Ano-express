```python
### MAIN
# a dict with gene families and their respective Pfam domain for extracting
gene_fams = {'COE': 'COesterase', 
             'OBP':'PBP_GOBP', 
             'OR':'7tm_6', 
             'Io':['Lig_chan','7tm_1'],
             'Gr': '7tm_7',
             'P450':'p450', 
             'Gst':['GST_N', 'GST_N_3', 'GST_C'], 
             'ABC':['ABC_membrane', 'ABC_tran'],
            'FAS':'ketoacyl-synt',
            'ELO':'ELO',
            'FAD':'FA_desaturase',
            'FAR':'NAD_binding_4'}


fc_data = pd.read_csv("../../results/fc_data.tsv", sep="\t")
pval_data = pd.read_csv("../../results/pval_data.tsv", sep="\t")
assert all(fc_data['GeneID'] == pval_data['GeneID'])

names = fc_data[['GeneID', 'GeneName']]
fc_data = fc_data.iloc[:, :-1]
pval_data.columns = pval_data.columns.str.replace("log2FoldChange", "padj")

genediff = {}

comps = fc_data.columns[1:-1]
for idx, comp2 in enumerate(comps):
    comp = comp2.replace("_log2FoldChange", "")

    df = fc_data[['GeneID', comp2, 'GeneName']]
    df.loc[:, 'padj'] = pval_data.loc[:, f'{comp}_padj']
    genediff[comp] = df
    #loop through gene fams 
    fam_dict = {}
    for fam, domain in gene_fams.items():
        gene_ids = gene_ids_from_domain(gene_annot_df, domain)
        fam_de = df.query("GeneID in @gene_ids").copy()

        fam_de.loc[:, 'de'] = np.where(
             np.logical_and(fam_de['padj'].between(0, 0.05, inclusive="both"), fam_de[comp2].ge(0)), 'Upregulated', 
             np.where(
                     np.logical_and(fam_de['padj'].between(0, 0.05, inclusive="both"), fam_de[comp2].le(0)), 'Downregulated', 'No Diff. expression'))

        fam_dict[fam] = fam_de
        genediff[comp].loc[:, 'Label'] = [id_ + " | " + name if name != "" else id_ for id_, name in zip(genediff[comp]['GeneID'].fillna(""), genediff[comp]['GeneName'].fillna(""))]
        
    genediff[comp] = dict(fam_dict)



sig_bool_df = pval_data.set_index("GeneID").iloc[:, :-2] > 0.05
fc_data = fc_data.set_index('GeneID').iloc[:, :-1]
fc_data[sig_bool_df.to_numpy()] = 0
fc_data = fc_data.fillna(0)

with PdfPages("../../results/genediff/GeneFamiliesHeatmap.pdf") as pdf:

    for fam, domain in gene_fams.items():

        ids = genediff[comp][fam]['GeneID'].to_numpy()


        fc_data2 = fc_data.reset_index()
        fc_data2.loc[:, 'GeneName'] = names['GeneName']
        fc_data2.loc[:, 'Label'] = [id_ + " | " + name if name != "" else id_ for id_, name in zip(fc_data2['GeneID'].fillna(""), fc_data2['GeneName'].fillna(""))]

        fc_data_fam = fc_data2.query("GeneID in @ids").set_index("GeneID")
        fc_data_labels = fc_data_fam['Label'].to_numpy()
        fc_data_fam = fc_data_fam.drop(columns=['Label', 'GeneName'])
        size = len(fc_data_fam)

        g = sns.clustermap(data=fc_data_fam, 
                    cmap="Blues", 
                    vmax=10, 
                    cbar=False,
                    linewidths=4,
                    col_cluster=False,
                    linecolor="white",
                    yticklabels=False, xticklabels=False,
                    tree_kws=dict(linewidths=3, colors="lightgray"),
                    figsize=[1, 1])
        ax = g.ax_heatmap
        ax.set_ylabel("")
        dendro = g.ax_row_dendrogram
        g.ax_cbar.set_visible(False)
        g.ax_row_dendrogram.set_visible(False) #suppress row dendrogram
        g.ax_col_dendrogram.set_visible(False) #suppress column dendrogram   

        order = g.dendrogram_row.reordered_ind
        fc_data_fam = fc_data_fam.iloc[order, :]
        fc_data_labels = fc_data_labels[order]

        fig, ax = plt.subplots(1, 4, figsize=[16,(size/3)], gridspec_kw={'width_ratios':[1, 2.6, 1, 1]})
        fig.suptitle(f"{fam} | pfam domains = {domain}", fontsize=18, fontweight='bold')
        dendro = scipy.cluster.hierarchy.dendrogram(g.dendrogram_row.linkage, no_plot=True,
                                                        above_threshold_color='lightgray', 
                                                        color_threshold=0, 
                                                        no_labels=True)
        ax[0] = plot_rotated_dendro(dendro, ax[0])   

        pal = sns.diverging_palette(120, 300, as_cmap=True)

        for i, species in enumerate(['gambiae', 'funestus', 'arabiensis']):
            i += 1 #skip first ax
            sp_bool = strains_df.eval("species == @species").values
            fc_data_fam_sp = fc_data_fam.loc[:, sp_bool]

            if fc_data_fam_sp.shape[1] < 2:
                continue

            sns.heatmap(data=fc_data_fam_sp, ax=ax[i], cmap=pal, cbar=False, linewidths=4, linecolor='white')
            ax[i].set_xticklabels(fc_data_fam_sp.columns.str.replace("_log2FoldChange", ""), fontsize=14, fontweight='bold')
            ax[i].set_xlabel(species, fontsize=14, fontweight='bold', ha='right', rotation_mode='anchor')
            ax[i].set_ylabel("")
            ax[i].set_yticklabels([])
            ax[i].set_yticks([])

        ax[-1].set_yticks(range(len(fc_data_labels)))
        ax[-1].set_yticklabels(labels=fc_data_labels, rotation=0,fontsize=14)
        ax[-1].tick_params(top=False,
                                bottom=False,
                                left=False,
                                right=False,
                                labelleft=False,
                                labelright=True,
                                labelbottom=True)
        plt.tight_layout()
        sns.despine(left=True, bottom=True, right=True)
        pdf.savefig(bbox_inches = 'tight')
```
