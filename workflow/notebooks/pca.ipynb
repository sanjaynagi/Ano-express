```python
normCounts = pd.read_csv("../../final_counts.txt", sep="\t")

normCounts = normCounts.set_index("GeneID")
normCounts = normCounts.iloc[:, 8:].copy().drop_duplicates()

normCounts = normCounts[normCounts.sum(axis=1) > 1000]
print(normCounts.shape)

metadata = pd.read_csv("../../config/ALLcoldata.txt", sep="\t")

normCounts = normCounts.T
normCounts.loc[:, 'species'] = metadata.loc[:, 'species'].to_numpy()

pca = PCA()
components = pca.fit_transform(normCounts.T.iloc[:-1,].astype(float).values)
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(
    components,
    #data_frame=normCounts,
    labels=labels,
    dimensions=range(4),
    #color=normCounts["species"]
)
fig.update_traces(diagonal_visible=False)
fig.show()
```
